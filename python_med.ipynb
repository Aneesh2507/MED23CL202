#List processing in python


Return a list were each element is a running sum.
sample input:[10,20,5,1]
sample output:[10,30,35,36]

#STEPS:-
1. create a empty result list.
2. keep a total variable starting at 0.
3. loopnthrough each number.
4. add to total,append total to result.

```python
def runningSum(nums):
  result=[]
  total=0
  for n in nums: #loop started
    total+= n    # add the value to total
    result.append(total) # append the sum to the result list
  return(result)
print(runningSum([10,20,30,2,-1]))
```

#Maximum number in a list

```python
def max(nums):
 max =nums[0] #assign first of list as maximum
 for n in nums:
   if n > max:
    max = n
 return max

print(max([10,2,3,5,7,9]))
```

#minimun of a list

```python
def min(nums):
 min =nums[0] #assign first of list as maximum
 for n in nums:
   if n < min:
    min = n
 return min

print(min([10,2,3,5,7,9]))
```

#Find two indices

```python
def twosum(nums,target):
  seen={}
  for i, n in enumerate(nums):
    diff = target - n
    if diff in seen:
      return [seen[diff],+i]
    seen[n] = i
    print(seen)
print(twosum([2,0,11,15],26))
```

#Experiment 4
Objective : To design a menu driven application for grocery store checkout system using list data structure in python.


```python
# Grocery Store Checkout
# Available products
products = {
    "Apples": 50,
    "Bananas": 30,
    "Milk": 60,
    "Bread": 40,
    "Eggs": 70
}

# Empty cart
cart = {}

print("\n===== Grocery Store Menu =====")
print("1. Show available products and prices")
print("2. Empty cart")
print("3. Display menu")
print("4. Add item to cart")
print("5. Remove item from cart")
print("6. Show cart")
print("7. Exit")

while True:
    choice = input("\nEnter your choice (1-7): ")

    if choice == "1":
        print("\n--- Available Products ---")
        for item in products:
            print(item, ":", products[item])

    elif choice == "2":
        cart.clear()
        print("\nCart is now empty!")

    elif choice == "3":
        print("\n===== Grocery Store Menu =====")
        print("1. Show available products and prices")
        print("2. Empty cart")
        print("3. Display menu")
        print("4. Add item to cart")
        print("5. Remove item from cart")
        print("6. Show cart")
        print("7. Exit")

    elif choice == "4":
        print("\n--- Available Products ---")
        for item in products:
            print(item, ":", products[item])
        item = input("Enter product name to add: ")

        if item in products:
            qty = int(input("Enter quantity: "))
            if item in cart:
                cart[item] = cart[item] + qty
            else:
                cart[item] = qty
            print(qty, item, "added to cart.")
        else:
            print("Item not available.")

    elif choice == "5":
        if cart == {}:
            print("\nCart is empty. Nothing to remove.")
        else:
            print("Your Cart:", cart)
            item = input("Enter product name to remove: ")
            if item in cart:
                qty = int(input("Enter quantity to remove: "))
                if qty >= cart[item]:
                    del cart[item]
                    print(item, "removed from cart.")
                else:
                    cart[item] = cart[item] - qty
                    print(qty, item, "removed from cart.")
            else:
                print("Item not in cart.")

    elif choice == "6":
        if cart == {}:
            print("\nYour cart is empty.")
        else:
            print("\n--- Your Cart ---")
            total = 0
            for item in cart:
                qty = cart[item]
                price = products[item] * qty
                total = total + price
                print(item, "x", qty, "=", price)
            print("Total Bill:", total)

    elif choice == "7":
        print("Thank you for shopping! Goodbye.")
        break

    else:
        print("Invalid choice, try again.")
```

# Nested Dictionary

```python
#CREATION OF NESTED DICTIONARY
student={
    "alice":{"math": 85,"science": 69},
    "bob":{"math": 75,"science": 90},
    "lanj":{"math": 85,"science": 99},
    "koj":{"math": 75,"science": 90},
    "don":{"math": 85,"science": 69},
}
student
```

#For data manipulation using dictionary convert the dictionary to dataframe

```python
import pandas as pd # pandas are library for data manipulation in python
data = pd.DataFrame(student)# pass the dictionary to dataframe
data
```

```python
data = pd.DataFrame(student).T # transpose the dataframe
data
```

Add

```python
data["average"] =data.mean(axis=1)
data
```

#sardha/just *for* fun

```python
data["Math+science"] = data["math"] + data["science"]
data
```

#Data Visualisation

```python
import matplotlib.pyplot as plt
#bar chart for student average performance
plt.figure(figsize=(8,5))
data["average"].plot(kind="bar",color="skyblue",edgecolor="black")
plt.title("average marks of students")
plt.xlabel("student")
plt.ylabel("average marks")
plt.show()
```

#How to visualise Subjectwise trend

```python
import matplotlib.pyplot as plt

# Bar Chart for student average performance
for subject in["math","science"]:
    plt.plot(data.index,data[subject],marker='o',label=subject)

plt.title("Student Average Performance")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
data[["math","science"]].plot(kind="box")
plt.title("Marks distribution by students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
