#List processing in python


Return a list were each element is a running sum.
sample input:[10,20,5,1]
sample output:[10,30,35,36]

#STEPS:-
1. create a empty result list.
2. keep a total variable starting at 0.
3. loopnthrough each number.
4. add to total,append total to result.

```python
def runningSum(nums):
  result=[]
  total=0
  for n in nums: #loop started
    total+= n    # add the value to total
    result.append(total) # append the sum to the result list
  return(result)
print(runningSum([10,20,30,2,-1]))
```

#count no.of even numbers in a list

```python
def even(nums):
 count = 0
 for n in nums:
   if n%2 == 0:
      count += 1
 return count
print(even([10,2,3,5,7,9]))
```

#Reverse a list

```python
def reverselist(nums):
  return nums[::-1]
print(reverselist([2,4,6,8,10]))
```

#Maximum number in a list

```python
def max(nums):
 max =nums[0] #assign first of list as maximum
 for n in nums:
   if n > max:
    max = n
 return max

print(max([10,2,3,5,7,9]))
```

#minimun of a list

```python
def min(nums):
 min =nums[0] #assign first of list as maximum
 for n in nums:
   if n < min:
    min = n
 return min

print(min([10,2,3,5,7,9]))
```

```python
def square(nums):
  return sorted(x**2 for x in nums)
print(square([1,2,3,4,5]))
```

#Find two indices

```python
def twosum(nums,target):
  seen={}
  for i, n in enumerate(nums):
    diff = target - n
    if diff in seen:
      return [seen[diff],+i]
    seen[n] = i
    print(seen)
print(twosum([2,0,11,15],26))
```
